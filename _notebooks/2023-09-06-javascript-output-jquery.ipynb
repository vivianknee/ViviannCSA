## Hacks
This lesson teaches you about tables.  In this hack, it is important that you analyze the difference in the styles of output shown.  

- Describe a benefit of a markdown table.
    - Readability: they are easy to write and understand. They use a straightforward syntax with pipes (|) and hyphens (-) to define rows, columns, and headers, making it simple for content creators.
    - Simplicity: Markdown tables are concise and uncluttered, which makes them suitable for quickly formatting data without the verbosity of HTML or spreadsheet applications. Users can quickly create tables without having to deal with extensive HTML tags or complex spreadsheet software.
    - Compatibility: Markdown is widely supported in various text editors, online platforms, and content management systems. 

- Describe the difference between HTML and JavaScript.
    - HTML is a markup language used for creating the structure and content of web pages. It consists of a set of predefined tags and elements that define the layout, headings, paragraphs, links, images, forms, and other structural components of a web page
    - JavaScript is a dynamic scripting language used to add interactivity and functionality to web pages. It allows developers to create interactive features, respond to user actions, and manipulate the content of web pages in real-time

- Describe a benefit of a table that uses JavaScript.
    - Data Visualization: JavaScript can be used to create interactive charts and graphs within tables, providing a visual representation of data alongside the tabular format. This helps users better understand complex data sets and trends.
    - Customization: JavaScript allows developers to customize the appearance and behavior of tables to match specific design requirements and user needs. This level of control can lead to a more user-friendly and tailored user experience.
    - Data Manipulation: JavaScript can facilitate data manipulation within tables, such as calculations, updates, and data validation. This is especially useful in applications where users need to input, modify, or work with data directly in the table.
