````java
import java.util.Scanner; //library for user input
import java.lang.Math; //library for random numbers
import java.util.Random;


public class horl {
    //variables
    private int targetNumber;
    private int numberOfGuesses;
    private boolean isGameOver;

    //initializing starting values 
    public horl() {
        Random random = new Random();
        targetNumber = random.nextInt(100) + 1; // Generate a random number between 1 and 100
        numberOfGuesses = 0;
        isGameOver = false;
    }

    public void play() {
        Scanner scanner = new Scanner(System.in);

        System.out.println("Welcome to the Higher Lower game!");
        System.out.println("Try to guess the target number between 1 and 100.");

        //while isGameOver is false, do this
        while (!isGameOver) {
            System.out.print("Enter your guess: ");
            int playerGuess = scanner.nextInt();
            numberOfGuesses++;

            if (playerGuess < targetNumber) {
                System.out.println("Higher!");
            } else if (playerGuess > targetNumber) {
                System.out.println("Lower!");
            } else {
                System.out.println("Congratulations! You've guessed the number " + targetNumber + " in " + numberOfGuesses + " guesses.");
                isGameOver = true;
            }
        }

        scanner.close();
    }

     //main method. calls play function
     public static void main(String[] args) {
         horl game = new horl();
         game.play();
     }
}

horl.main(null);
````

````java
import java.util.Scanner; //library for user input
import java.lang.Math; //library for random numbers
import java.util.HashMap; //stores data in (Key, Value) pairs, can access them by an index of another type
import java.util.Map; //represents a mapping between a key and a value. More specifically, a Java Map can store pairs of keys and values

public class ticTacToe {

    //instance variables. char is a data type that is used to store a single character
    private char[][] board; //2D char array that represents the game board.
    private char currentPlayer; //whether the player is X or O
    private Map<Character, String> players; //makes letter and player a key and value pair. ex: X, Player 1

    public ticTacToe() {
        board = new char[3][3]; //3x3 board
        currentPlayer = 'X'; //sets player to X
        players = new HashMap<>();
        //accessing this data with HashMap
        players.put('X', "Player 1"); //equates X to player 1
        players.put('O', "Player 2"); //equates O to player 2
        initializeBoard(); //clear board
    }

    public void play() {
        Scanner scanner = new Scanner(System.in);
        System.out.println("Welcome to Tic Tac Toe!");

        while (true) {
            //generate board
            displayBoard();

            //player is prompted to enter a row and column for their move, and the input is validated (see if theres alr an X or O there)
            int row, col;

            do {
                System.out.print(players.get(currentPlayer) + ", enter row (1-3) and column (1-3): ");
                row = scanner.nextInt() - 1;
                col = scanner.nextInt() - 1;
            } while (!isValidMove(row, col));

            //takes values that player entered and uses it to validate move, put move on the board, etc
            makeMove(row, col);
            
            //running checkWin and checkDraw functions and outputs result
            if (checkWin()) {
                displayBoard();
                System.out.println(players.get(currentPlayer) + " wins!");
                break;
            } else if (checkDraw()) {
                displayBoard();
                System.out.println("It's a draw!");
                break;
            }

            currentPlayer = (currentPlayer == 'X') ? 'O' : 'X';
        }

        scanner.close();
    }

    //clears the board each game
    private void initializeBoard() {
        for (int i = 0; i < 3; i++) { //rows
            for (int j = 0; j < 3; j++) { //columns
                board[i][j] = '-'; //fill each with this
            }
        }
    }

    //what the board looks like
    private void displayBoard() {
        System.out.println("-------------");
        for (int i = 0; i < 3; i++) {
            System.out.print("| ");
            for (int j = 0; j < 3; j++) {
                System.out.print(board[i][j] + " | ");
            }
            System.out.println();
            System.out.println("-------------");
        }
    }

    private boolean isValidMove(int row, int col) {
        if (row < 0 || row >= 3 || col < 0 || col >= 3) {
            System.out.println("Invalid move. Row and column must be between 1 and 3.");
            return false;
        }
        if (board[row][col] != '-') {
            System.out.println("Invalid move. Cell is already occupied.");
            return false;
        }
        return true;
    }

    //places X or O in right spot
    private void makeMove(int row, int col) {
        board[row][col] = currentPlayer;
    }

    //checking all combinations of possible wins
    private boolean checkWin() {
        //changing row index, checks for vertical wins
        for (int i = 0; i < 3; i++) { 
            if (board[i][0] == currentPlayer && board[i][1] == currentPlayer && board[i][2] == currentPlayer) {
                return true;
            }
            //changing col index each time, checks for horizontal wins
            if (board[0][i] == currentPlayer && board[1][i] == currentPlayer && board[2][i] == currentPlayer) {
                return true;
            }
        }

        //check row, col for diagonal wins
        if (board[0][0] == currentPlayer && board[1][1] == currentPlayer && board[2][2] == currentPlayer) {
            return true;
        }
        if (board[0][2] == currentPlayer && board[1][1] == currentPlayer && board[2][0] == currentPlayer) {
            return true;
        }
        return false;
    }

    private boolean checkDraw() {
        for (int i = 0; i < 3; i++) {
            for (int j = 0; j < 3; j++) {
                if (board[i][j] == '-') {
                    return false; // if an empty cell found, game is not a draw
                }
            }
        }
        return true; // if all cells are filled, game is a draw
    }

    public static void main(String[] args) {
        ticTacToe game = new ticTacToe();
        game.play();
    }
}

ticTacToe.main(null);
````

````java
import java.util.Scanner; //library for user input
import java.lang.Math; //library for random numbers
import java.util.Random;

public class rps {
    //defines an enumeration which is a special data type which contains a set of predefined constants. constants never change while variables do
    private enum Move {
        R, P, S
    }

    //instance variables
    private Move playerMove;
    private Move computerMove;
    private int playerScore;
    private int computerScore;

    //defines variables
    public rps() {
        playerScore = 0;
        computerScore = 0;
    }

    public void play() {
        Scanner scanner = new Scanner(System.in);
        Random random = new Random();

        System.out.println("Welcome to Rock Paper Scissors!");

        while (true) {
            System.out.print("Enter your move (rock(r), paper(p), scissors(s), or quit(q)): ");
            String input = scanner.nextLine().trim().toLowerCase();

            if (input.equals("q")) {
                break;
            }

            if (!input.equals("r") && !input.equals("p") && !input.equals("s")) {
                System.out.println("Invalid move. Please choose r, p, or s.");
                continue;
            }

            //player move is assigned based on the input. 
            playerMove = Move.valueOf(input.toUpperCase());
            //computer move is generated randomly using the nextInt(3) method, which generates a random integer from 0 to 2.
            computerMove = Move.values()[random.nextInt(3)];

            System.out.println("Computer chose: " + computerMove);

            int result = determineWinner(playerMove, computerMove);

            if (result == 0) {
                System.out.println("It's a tie!");
            } else if (result == 1) {
                System.out.println("You win!");
                playerScore++;
            } else {
                System.out.println("Computer wins!");
                computerScore++;
            }

            System.out.println("Your score: " + playerScore + " | Computer score: " + computerScore);
        }

        System.out.println("Thanks for playing!");
        scanner.close();
    }

    private int determineWinner(Move playerMove, Move computerMove) {
        if (playerMove == computerMove) {
            return 0; // tie
        } else if ((playerMove == Move.R && computerMove == Move.S)
                || (playerMove == Move.P && computerMove == Move.R)
                || (playerMove == Move.S && computerMove == Move.P)) {
            return 1; // player wins
        } else {
            return -1; // computer wins
        }
    }

    public static void main(String[] args) {
        rps game = new rps();
        game.play();
    }
}

rps.main(null);
````
